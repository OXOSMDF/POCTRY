```python
import requests
from bs4 import BeautifulSoup
import spacy
```

```python
nlp = spacy.load("en_core_web_sm")
```

```python
def fetch_news(query): 
    url = f"https://news.google.com/rss/search?q={query}&hl=en-DE&gl=DE&ceid=DE:en" 
    response = requests.get(url) 
    soup = BeautifulSoup(response.content, 'xml') 
    articles = soup.find_all('item') 
    return [(article.title.text, article.link.text, article.pubDate.text) for article in articles] 
```

```python
def extract_information(text): 
    doc = nlp(text) 
    companies = [ent.text for ent in doc.ents if ent.label_ == "ORG"] 
    amounts = [ent.text for ent in doc.ents if ent.label_ == "MONEY"] 
    return companies, amounts
```

```python
articles = fetch_news("merger acquisition Germany") 
# Extract and display information from articles 
for title, link, pubDate in articles: 
    companies, amounts = extract_information(title) 
    print(f"Title: {title}") 
    print(f"Link: {link}") 
    print(f"Published Date: {pubDate}") 
    print(f"Companies: {companies}") 
    print(f"Amounts: {amounts}") 
    print("\n" + "-"*80 + "\n")
```

```python
def check_merger(title):
    doc = nlp(title)
    for token in doc:
        if token.text.lower() in ["merges", "acquires", "buys"]:
            subj = [child for child in token.children if child.dep_ in ["nsubj", "nsubjpass"]]
            obj = [child for child in token.children if child.dep_ in ["dobj", "pobj"]]
            if subj and subj[0].ent_type_ == "ORG" and obj and obj[0].ent_type_ == "ORG":
                return f"{subj[0].text} merges {obj[0].text}"
    return "No merger found in the title."

title = "Company A acquires Company B"
print(check_merger(title))
```
